```python
import backtrader as bt
import datetime as dt
import datetime
import matplotlib
import math
import talib
import numpy as np
%matplotlib inline
import backtrader.analyzers as btanalyzers
```

```python
import logging

logging.basicConfig(filename="log.txt", level=logging.DEBUG)
```

```python
class RsiSignalStrategy(bt.Strategy):
    params = dict(rsi_periods=14,rsi_upper=70,rsi_lower=30,rsi_mid=50)

    def __init__(self):
        
        self.sma = bt.indicators.SimpleMovingAverage(
            self.datas[0], period=2)

        self.rsi = bt.indicators.RSI(
            self.datas[0])
        
        
    def log(self, txt, dt=None):
        """ Logging function fot this strategy"""
        dt = dt or self.datas[0].datetime.date(0)
        logging.debug('%s, %s' % (dt.isoformat(), txt))
        
        
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log("Buy EXECUTED {}".format(order.executed.price))
            elif order.issell():
                self.log("Sell EXECUTED {}".format(order.executed.price))

            self.bar_executed = len(self)

        self.order = None
    
    def next(self):
        
        if self.data_close[0] > self.sma[0] and self.rsi[0] > 60:

            if self.position.size >= 0:
                self.log("BUY CREATE A, '%.2f', '%.0f'" % (self.data_close[0], self.position.size))
                self.order = self.buy(size=1)

        if self.data_close[0] < self.sma[0] and self.rsi[0] < 50:

            if self.position.size > 0:
                self.log("SELL CREATE B, '%.2f', '%.0f'" % (self.data_close[0], self.position.size))
                self.order = self.sell(size=abs(self.position.size))

        elif self.data_close[0] < self.sma[0]:

            if self.position.size > 0:
                self.log("SELL CREATE B, '%.2f', '%.0f'" % (self.data_close[0], self.position.size))
                self.order = self.sell(size=abs(self.position.size))
```

```python
cerebro = bt.Cerebro()
cerebro.broker.setcash(1000.0)

data = bt.feeds.GenericCSVData(dataname="../DATA/SBIN.csv",
                               timeframe=bt.TimeFrame.Minutes,
                               fromdate = dt.datetime(2018, 3, 1), 
                               todate=dt.datetime(2019, 3, 1),
                               open_interest = -1,
                               adj_close = -1)

cerebro.adddata(data)
cerebro.addstrategy(RsiSignalStrategy)
cerebro.addanalyzer(btanalyzers.SharpeRatio, _name='mysharpe')
cerebro.addanalyzer(btanalyzers.PyFolio, _name='pyfolio')
    
print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())

thestrats = cerebro.run()
print(cerebro.broker.get_value()-1000)
```

```python
thestrat = thestrats[0]

strat = thestrat
pyfoliozer = strat.analyzers.getbyname('pyfolio')
returns, positions, transactions, gross_lev = pyfoliozer.get_pf_items()

print('Sharpe Ratio:', thestrat.analyzers.mysharpe.get_analysis())
print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())

cerebro.plot()
```

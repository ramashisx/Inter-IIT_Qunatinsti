# PIVOT POINT

Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
```

Import Data 

```python
df2 = pd.read_csv(r'SBIN.csv')
```

```python
df2 = df2[:100]
```

```python
def pv(_open, high, low, close, prev_open, prev_close):
    close = (_open + high + low + close) / 4
    _open = (prev_open + prev_close) / 2
    high = max([high, _open, close])
    low = min([low, _open, close])
    pp = (high+low+close)/3
    s1 = 2*pp - high
    r1 = 2*pp - low
    return s1, r1
```

```python
prev_close = df2["Close"].iloc[0]
prev_open = df2["Open"].iloc[0]
s = []
r = []
for i in range(0, len(df2)):
    a, b = pv(df2["Open"].iloc[i], df2["High"].iloc[i], df2["Low"].iloc[i], df2["Close"].iloc[i], prev_open, prev_close)
    prev_open = df2["Open"].iloc[i]
    prev_close = df2["Close"].iloc[i]
    s.append(a)
    r.append(b)
```

```python
df2["s"] = s
df2["r"] = r
```

```python
ohlc= df2[['Date', 'Close', 'High', 'Low','Open']].copy()
f1, ax = plt.subplots(figsize = (40, 20))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.plot(df2['Date'], df2['Close'], label = 'Price Level')
plt.plot(df2['Date'], df2['High'], label = 'Price Level')
plt.plot(df2['Date'], df2['Low'], label = 'Price Level')
plt.plot(df2['Date'], df2['r'], label = 'Resistance One')
plt.plot(df2['Date'], df2['s'], label = 'Support One')
plt.legend(loc=2, prop={'size': 10})
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Close')
```

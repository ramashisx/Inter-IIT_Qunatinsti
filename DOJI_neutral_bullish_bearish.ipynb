```python
import yfinance as yf 
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import talib as ta

df = pd.read_csv('hdfc.csv') #works with any df with ohlc columns
```

## support line calculations

```python
def isSupport(df,i):
  support = df['Low'][i] < df['Low'][i-1]  and df['Low'][i] < df['Low'][i+1] and df['Low'][i+1] < df['Low'][i+2] and df['Low'][i-1] < df['Low'][i-2]
  return support
def isResistance(df,i):
  resistance = df['High'][i] > df['High'][i-1]  and df['High'][i] > df['High'][i+1] and df['High'][i+1] > df['High'][i+2] and df['High'][i-1] > df['High'][i-2]
  return resistance

levels = []
for i in range(2,df.shape[0]-2):
  if isSupport(df,i):
    levels.append((i,df['Low'][i]))
  elif isResistance(df,i):
    levels.append((i,df['High'][i]))

from sklearn.cluster import KMeans

def clustering(levels):
    b=[]
    for a in levels:
        b.append(a[1])

    km = KMeans()
    km.fit(np.array(b).reshape(-1,1))
    return km.cluster_centers_

centres = clustering(levels)
```

## support lines w different patterns

```python
doji = (ta.CDLDOJI(df.Open,df.High,df.Low,df.Close)) 
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
for aaa in centres:
    fig.add_hline(y=aaa[0])
    
for i in range(len(doji)):
    if(doji[i]!=0):
        cprice = df.Close[i]
        prevprice = df.Close[i-1]
        nextprice = df.Close[i+1]
        for x in centres:
            if(1.005*cprice>x[0] and 0.995*cprice<x[0] and ((prevprice<x[0] and nextprice<x[0]) or (prevprice>x[0] and nextprice>x[0]))):
                fig.add_vline(x=df.Date[i])
                break
        
fig.show()
```

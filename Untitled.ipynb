```python
def opt(sl,tp):
    trade = 'None'
    stoploss = 0
    takeprofit = 0
    price = 0
    goodorbad = []
    for i in range(len(data)-1):
        if(trade == 'None'):
            if((data.Open[i+1] - data.High[i])/data.Close[i]>0.0025):
                trade = 'Long'
                price = data.Open[i+1]
                stoploss = data.Open[i+1]*(1-sl)
                takeprofit = data.Open[i+1]*(1+tp)
            elif((data.Low[i] - data.Open[i+1])/data.Close[i]>0.0025):
                price = data.Open[i+1]
                trade = 'Short'
                takeprofit = data.Open[i+1]*(1-tp)
                stoploss = data.Open[i+1]*(1+sl)
        if(trade =='Long'):
            if(stoploss>data.Low[i+1]):
                goodorbad.append(stoploss-price)
                trade = 'None'
            elif(data.High[i+1]>takeprofit):
                goodorbad.append(takeprofit-price)
                trade = 'None'
        if(trade == 'Short'):
            if(stoploss<data.High[i+1]):
                goodorbad.append(price-stoploss)
                trade = 'None'
            elif(data.Low[i+1]<takeprofit):
                goodorbad.append(price-takeprofit)
                trade = 'None'
    goodorbad = np.array(goodorbad)*50
    return goodorbad.sum()/(goodorbad.std()*np.sqrt(250))
```

```python
def gapanalysis(data,daysin):
    gap = 0
    x=0
    gapsup=0
    gapsdown=0
    gapupnotfilled=0
    gapupfilled=0
    gapdownnotfilled=0
    gapdownfilled=0
    for i in range(len(data)-1):
        cond1 = (data.Open[i+1] - data.Close[i])/data.Close[i]>0.0
        cond2 = (data.Open[i+1] - data.Close[i])/data.Close[i]<0.005
        if(gap==0 and (cond1 or cond2)):
            gap = data.Close[i]
            gapsup+=1
            if(cond2):
                gapsup-=1
                gapsdown+=1
                gap = -data.Close[i]
#             print('Gap Identified!')
            x=0
        elif(gap!=0):
            if(gap>0):
                if(data.Close[i]<gap):
                    x=0
    #                 print('Gap filled!')
                    gap=0
                    gapupfilled+=1
                elif(x==daysin):
    #                 print('Gap not filled')
                    gap=0
                    gapupnotfilled+=1
                else:
                    x+=1
            else:
                if(-data.Close[i]<gap):
                    x=0
    #                 print('Gap filled!')
                    gap=0
                    gapdownfilled+=1
                elif(x==daysin):
    #                 print('Gap not filled')
                    gap=0
                    gapdownnotfilled+=1
                else:
                    x+=1
    return gapsup,gapsdown,gapupfilled,gapupnotfilled,gapdownfilled,gapdownnotfilled
```

```python
gapanalysis(pd.read_csv('NIFTY.csv'),5)
```
